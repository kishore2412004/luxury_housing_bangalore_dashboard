```python
import mysql.connector

connection = mysql.connector.connect(
    host = "localhost",
    user = "root",
    password = "kishore24",
    database = "luxury_housing"
)

cursor = connection.cursor()

def creation_of_table(cursor, table_name, df):
    dtype_map ={
        "int64": "int",
        "float64": "float",
        "object": "varchar(255)",
        "bool": "boolean",
        "datetime64[ns]": "datetime"

    }

    columns = []
    for cols in df.columns:
        dtype = str(df[cols].dtype)
        sql_type= dtype_map.get(dtype, "VARCHAR(255)")
        columns.append(f"`{cols}` {sql_type}")


    table_declaration = ", ".join(columns)
    query = f"CREATE TABLE `{table_name}` ({table_declaration});"

    cursor.execute(query)
    connection.commit()


def insertion_of_table(cursor, connection, df, table_name):
    col= ", ".join([f"`{cols}`" for cols in df.columns])
    placeholders = ", ".join(["%s"]* len(df.columns))
    query = f"INSERT INTO `{table_name}`({col}) VALUES ({placeholders});"

    rows = list(df.itertuples(index=False, name= None))
    cursor.executemany(query, rows)
    connection.commit()



table_name = "luxury_housing"
creation_of_table(cursor, table_name, luxury_housing_db)
insertion_of_table(cursor, connection, luxury_housing_db, table_name)
print(f"Inserted {len(luxury_housing_db)} rows into `{table_name}`")




cursor.close()
connection.close()

```
